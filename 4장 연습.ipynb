```python
import pandas as pd
import matplotlib.pyplot as plt
drink = pd.read_csv('C:/Users/pc/Desktop/판다스터디/drinks.csv')
drink = drink.fillna('NA')
drink
```

```python
#1.대륙별 평균 wine_servings 를 출력하기
drink.groupby('continent')['wine_servings'].mean() 
dr = drink.groupby('continent')['wine_servings'].mean()
dr.plot(kind = 'bar', figsize = (20,10), width = 0.5, color = ['red','blue','orange','yellow','black','green'])
plt.title('wine servings',fontsize = 30)  #autopct = '%1.1f%%'
plt.show()
```

```python

def program():
    scores = list(map(int,input("점수를 입력하십시오 : ").split(',')))
    k = int(input("3에서 100까지의 정수를 입력하십시오:"))
    best_k = []
    for i in scores:
          a = max(scores)
          if len(best_k) < k:
            best_k.append(a)
            scores.remove(a)
            low_scores = min(best_k)
            print(low_scores)
          else:
              best_k.append(a)
              for x in best_k:
                if a > x:
                    best_k.remove(min(best_k))
                    best_k.append(a)
              print(low_scores)
print(program())
```
